# Ordinære differensialligninger

```python
# @title Import av moduler
import numpy as np
import matplotlib.pyplot as plt
```

Her ser vi på førsteordens differensialligninger på formen
$$x'(t) = f(t,x(t))$$
for en funksjon 
$$
\begin{eqnarray}
\mathbb{R}^2 &\xrightarrow f& \mathbb{R}, \\*
(t, x) &\mapsto& f(t, x).
\end{eqnarray}
$$

## Eksempel

La oss se på tilfellet der $f(t,x) = 2$ for alle $x$ og $t$
slik at 
$$x'(t) = 2$$
Vi vet at funksjonen $x_0(t) = 2t$ har $x_0'(t) = 2$. Videre vet vi at dersom $g'(t) = 2$ for all $t$, da er $x_0'(t) - g'(t) = 0$ for alle $t$.
Derfor er $x_0(t) - g(t) = C$ der $C$ er en konstant. Vi kan konkludere at alle løsninger til $x'(t) = 2$ er på formen $x(t) = x_C(t) = 2t + C$ for en konstant $C$.

```python
t = np.linspace(-2,2,50)
for i in range(-2,3):
    plt.plot(t, 2*t + i, label=f'C = {i}')
plt.legend()
```

Hvis vi har
$$x'(t) = 2, \qquad x(0) = 0$$
da er det bare løsningen $x_0$ som oppfylder andre betingelse, altså $x_0(0) = 0$.

## Eksempel

Løs differensialligningen
$$x'(t) = x(t), \quad \text{med} \quad x(0) = 2$$

Her kjenner vi en løsning til $x'(t) = x(t)$, nemlig $x_1(t) = e^t$. Bemerk at $x_1(0) = 1$, så $x_1(x)$ er ikke løsningen vi er på jakt etter. Hvis $g(t)$ er en annen løsning, da gir regelen for derivasjon av brøker at
$$ \left(\frac{g(t)}{x_1(t)}\right)' = 0$$

## Oppgave:

Sjekk påstanden over.

Det vil si at funksjoen $\frac{g(1)}{x_1(1)}$ er konstant, så $g(1) = C x_1(t) = Ce^t$ for en konstant $C$. Skriver vi $x_C(t) = C e^t$, så er $x_C(0) = C$, Derfor er $x_2(t) = 2e^t$ 
en funksjon som oppfylder de gitte betingelsene:
$$x_2'(t) = x_2(t), \quad \text{med} \quad x_2(0) = 2$$.

# Retningsfelt

Førsteordens differensialligninger på formen
$$x'(t) = f(t,x(t))$$
kan tolkes grafisk ved hjelp av såkalte retningsfelt.
La $x(t)$ være en løsning til diffligningen over, og se på parametriseringen $t \mapsto (t, x(t))$ for grafen til funksjonen $x(t)$. Tænker vi på
dette som banen til en partikkel som beveger seg i planet, og tenker vi på $t$ som tid, da er hastighetsvektoren til tidspunktet $t$ git ved $(1,x'(t)) = (1, f(t,x(t)))$. 
Siden patikkelen til tiden $t$ befinner seg i posisjon $(t, x(t))$ kan vi grafisk angi hastighetsvektoren ved å tegne en pil fra punktet i $\mathbb{R}^2$ med koordinater $(t, x(t))$ 
til punktet med koordinater $(t, x(t)) + (1, f(t,x(t))) = (t+1, x(t) + f(t, x(t)))$.


## Oppgave:

La $a > 0$, og se på parametriseringen $t \mapsto (at, x(at))$ for grafen til funksjonen $x(t)$. Hvordan kan vi grafisk angi hastighetsvektoren for en partikkel som beveger seg på denne måten?

## Eksempel

La $f(t,x) = x$ slik at vi får diffligningen $x'(t) = x(t)$. Vi ser på løsningen $x_2(t) = 2e^t$ som vi fant over. For $t = 2$ er $f(t, x(t)) = x(t) = x(2) = 2 e^2$. Vi vil altså tegne en pil som begynner i punktet $(t, x(t)) = (2, x(2)) = (2, 2e^2)$ og som ender i punktet $(t+1, x(t) + f(t, x(t))) = (1 + 2, 2e^2 + 2e^2) = (3, 4e^2)$. Se figuren under. Koden kan du ignorere.

```python
# @title Plotting av graf med pil
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Definer funksjonen x(t) = 2e^t
def x_2(t):
  return 2 * np.exp(t)

# Definer funksjonen f(t, x) = x
def f(t, x):
  return x

# Startpunkt for pilen
t_start = 2
x_start = x_2(t_start)

# Sluttpunkt for pilen
t_end = t_start + 1
x_end = x_start + f(t_start, x_start)

t = np.linspace(0, 3, 100)


# Lag en FancyArrowPatch
plt.figure(figsize=(8,6))
arrow = patches.FancyArrowPatch((t_start, x_start), (t_end, x_end),
                                 mutation_scale=20,
                                 #arrowstyle="->",
                                 color="k")

# Legg til pilen til plottet
ax = plt.gca()
ax.add_patch(arrow)
ax.plot(t, x_2(t))

# Sett grenser for aksene for å vise pilen
# ax.set_xlim([0, 3.2])  
# ax.set_ylim([0, 40]) 

# Sett akser og tittel
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.set_title('Grafen til $x(t)$ og pil fra $(2, 2e^2)$ til $(3, 4e^2)$');
```

```python
# @title Plotting av graf med mange piler
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Definer funksjonen x_2(t) = 2e^t
def x_2(t):
  return 2 * np.exp(t)

# Definer funksjonen f(t, x) = x
def f(t, x):
  return x

# Startpunkt for pilen
t_start = 2
x_start = x_2(t_start)

# Sluttpunkt for pilen
t_end = t_start + 1
x_end = x_start + f(t_start, x_start)

t = np.linspace(0, 3, 100)


# Lag x-verdier for pilene
t_piler = np.linspace(0, 3, 18)  # 16 piler jevnt fordelt

# Plot grafen
plt.plot(t, x_2(t), label='x(t) = 2e^t')

# Pilfaktor
a = 0.2  # Juster pilens vannrette lengde her

# Tegn piler
for t_start in t_piler:
    x_start = x_2(t_start)
    t_end = t_start + a 
    x_end = x_start + f(t_start, x_start) * a

    arrow = patches.FancyArrowPatch((t_start, x_start), (t_end, x_end),
                                     mutation_scale=20,  # Juster pilhodestørrelse
                                     #arrowstyle="->",
                                     color="k")
    plt.gca().add_patch(arrow)

# Sett akser, tittel og legend
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Grafen til x(t) = 2e^t med tangentpiler')
plt.legend()

# Vis plottet
plt.show()
```

## Grov beskrivelse av Eulers metode
Vi kan nå snu på det og si at hvis vi vil løse diffligningen 
$$x'(t) = x(t), \quad \text{med} \quad x(0) = 2$$
da kan vi starte til venstre i punktet $(0, 2)$ og ta et steg med pilen som begynner der. Da havner vi omtrent på den blå kurven, og vi kan gå langs en pil som begynner der vi havner, og så videre. Dette er en grov beskivelse av Eulers metode for løsning av diffligninger som også er illustrert under.

```python
# @title På vei mot Eulers metode
import numpy as np
import matplotlib.pyplot as plt

# Definer funksjonen x_2(t) = 2e^t
def x_2(t):
  return 2 * np.exp(t)

# Definer funksjonen f(t, x) = x
def f(t, x):
  return x

n = 20


# Definerer x- og y-verdiene
t = np.linspace(0, 3, num=n+1)
x = np.linspace(0, 40, num=n+1)

# Lager et firkantet gitter av de oppgitte x- og y-verdiene
T, X = np.meshgrid(t, x)

# Definerer f med hensyn på T og X
F = f(T, X)

# Definerer "steget" i t- og x-retning
dt = np.ones_like(T)
dx = F

# plt.quiver tar for seg å lage alle tangentpilene [dt, dx] i alle punktene som opptrer i gitteret vi lagde.
plt.quiver(T, X, dt, dx, scale=500, scale_units='width', color='black', angles='xy')
plt.plot(t, x_2(t), label='x(t) = 2e^t')
plt.ylim((0, 50))
plt.grid()
plt.show()
```

# Eulers metode

Gitt en diffligning
$$x'(t) = f(t,x(t)) \quad \text{med} \quad x(t_0) = x_0$$
og jevnt fordelte $t$-verdier $t_0 < t_1 < t_2 < \dots < t_n$ med steglengde $\Delta t = t_1 - t_0$
vil vi finne en tilnærming av verdiene $x_1 = x(t_1), \dots, x_n = x(t_n)$
til løsningen $x(t)$ med $x(t_0) = x_0$.

Vi starter med å la $x_1 = x_0 + \Delta t \cdot f(t_0, x_0)$, og fortsetter med å la $x_2 = x_1 + \Delta t \cdot f(t_1, x_1)$. 
Hvis vi har beregnet tilnærminger frem til en tilnærming $x_i$ til $x(t_i)$ lar vi $x_{i+1} = x_i + \Delta t \cdot f(t_i, x_i)$.

![eulersmetode](Euler_method.svg)
[Public Domain](https://commons.wikimedia.org/w/index.php?curid=2143753)

```python
import numpy as np

def euler_method(f, t0, x0, tn, n):
  """
  Bruker Eulers metode for å løse differensialligningen x' = f(t, x).

  Argumenter:
    f: Funksjonen som representerer høyresiden av differensialligningen.
    t0: Startverdien for t.
    x0: Startverdien for x.
    tn: Sluttverdien for t.
    n: Antall steg.

  Returnerer:
    En NumPy-array som inneholder de omtrentlige verdiene av x ved hvert steg.
  """

  t_values = np.linspace(t0, tn, n + 1)
  x_values = [x0]
  delta_t = t_values[1] - t_values[0]

  for t_val in t_values[:-1]:
    x_next = x_values[-1] + delta_t * f(t_val, x_values[-1])
    x_values.append(x_next)

  return t_values, np.array(x_values)
```

```python
# Define the function f(t, x) = x
def f(t, x):
  return x
```

```python
# Seter initial betingelser og parametre
t0 = 0   # Første x-verdi
x0 = 2  # Iinital y-verdi
tn = 3  # Siste x-verdi
n = 4   # Antall steg

# Anvend Eulers metode
t_values, x_values = euler_method(f, t0, x0, tn, n)
```

```python
# Vi plotter resultatet, med prikker for tallene vi beregnet i Eulers metode 
# sammen med den eksakte løsningen til diffligningen.
plt.plot(t_values, x_values)
plt.scatter(t_values, x_values)
t = np.linspace(t0, tn, 100)
plt.plot(t, 2*np.exp(t))
plt.xlabel('t')
plt.ylabel('x')
plt.title(r"Eulers metode for x′(t)=x(t) i fire steg");
```

```python
# @title Plot med retningsdiagram og steg av metoden til Euler
# Definer funksjonen y(x) = 2e^x
def x_2(t):
  return 2 * np.exp(t)
    
N = 20

# Definerer t- og x-verdiene
t = np.linspace(0, 3, num=N+1)
x = np.linspace(0, 40, num=N+1)

# Setter startbetingelsene og parameterne
t0 = t[0]   # Første t-verdi
x0 = 2      # Initial x-verdi
tn = t[-1]  # Siste t-verdi
n = 3       # Antall steg for Eulers metode

# Anvender Eulers metode
t_values, x_values = euler_method(f, t0, x0, tn, n)

# Lager et firkantet gitter av de oppgitte x- og y-verdiene
T, X = np.meshgrid(t, x)

# Definerer f med hensyn på T og X
F = f(T, X)

# Definerer "steget" i x- og y-retning
dt = np.ones_like(T)
dx = F

# plt.quiver lager tangentpiler [dt, dx] i alle punktene i gitteret vi har laget.
plt.quiver(T, X, dt, dx, scale=500, scale_units='width', color='black', angles='xy')
plt.plot(t_values, x_values, '-o', label=r'Tilnærmet løsning')
tt = np.linspace(t0, tn, 100)
plt.plot(tt, 2*np.exp(tt), label=r'Eksakt løsning')
# plt.ylim((0, 50))
# plt.grid()
plt.xlabel('t')
plt.ylabel('x')
plt.legend(loc='upper left')
plt.title(r"Eulers metode for x′(t)=x(t) i " + f"{n} steg");
```

## Eksempel 4.2.2 (Torricellis lov, (Eksempel 12.2.1 og 12.2.5 i MIP)):
![http://caetool.com/2019/06/13/p0088/](Toricelli2.png "Toricelli")

Toricellis lov sier at når væske strømmer ut av et kar gjennom en liten åpning, så er hastigheten, $v$,
i utløpet gitt ved $v = \sqrt{2gh}$, der $h$ er høydeforskjellen mellom væskeoverflaten og åpningen, og $g$ er
tyngdeakselerasjonen. I dette eksempelet skal vi se på hvordan vi kan bruke dette til å bestemme høyden
$h(t)$ som en funksjon av tiden.

Vi antar at ved starten ($t = 0$), så er væskeoverflaten en høyde $h_0$ over åpningen i karet. Vi trekker
så ut en propp, slik at væsken strømmer ut av åpningen. I starten vil væsken strømme raskt ut, men
hastigheten vil avta etter hvert som høyden blir mindre. Ved hjelp av Torricellis lov kan vi sette opp en
differensialligning som beskriver problemet.

Anta at karet vi ser på er en sylinder med grunnflate $A$, og at åpningen i bunnen av karet har areal $B$.
Volumet av væsken, når hastigheten er $v(t)$, som slipper ut av karet i løpet
av et lite tidsrom, $\Delta t$, er
omtrent $Bv(t)\Delta t$. Samtidig som væske strømmer ut av karet, vil volumet som er igjen i karet avta fra
$Ah(t)$ til $Ah(t + ∆t)$. Forskjellen her må altså være
$$
Ah(t) − Ah(t + ∆t) \approx Bv(t)∆t.
$$

Hvis vi nå deler på $\Delta t$ og lar $\Delta t \to 0$, får vi
$$
\lim_{\Delta t \to 0} \frac{Ah(t) − Ah(t + \Delta t)}{\Delta t} = Bv(t). 
$$
Venstresiden av ligningen over, gjenkjenner vi som den deriverte av $−Ah(t)$, og vi har dermed at
$−Ah'(t) = Bv(t)$.
Løser vi så for $h'(t)$, får vi til slutt at Torricellis lov gir oss
$$
h'(t) = -\frac{B}{A} v(t) = -\frac{B}{A} \sqrt{2gh(t)}
$$
Sriver vi $x(t) = h(t)$ og $f(t, x) = -\frac{B}{A} \sqrt{2gx(t)}$ ser vi at dette er en differensialligning
på formen $x'(t) = f(x, t)$.

Når vi nå skal løse dette problemet, må vi spesifisere størrelsene som inngår. Vi kan for eksempel la
grunnflaten til sylinder ha areal $A = 1m^2$ , og åpningen ha areal
$B = 10{\mathrm{cm}}^2  = 0.001 m^2 = 10^{−3} m^2$.
Videre kan vi la starthøyden på vannflaten være $h_0 = 1m$. Dessuten vet vi at $g \approx 9.81 m / s^2$.
Ignorerer vi fysiske enheter,
ser ligningen vår ut slik
$h'(t) = - 0.001 \sqrt{2 \cdot 9.81 \cdot h(t)}$
eller
$x'(t) = f(t,x)$ der
$f(t,x) = - 0.001 \sqrt{2 \cdot 9.81 \cdot x(t)}$

For å løse problemet numerisk med Eulers metode, må vi også velge en skrittlengde, $\Delta t$.
Her kan vi prøve oss frem inntil resultatet ser ut til ikke å endre seg når skrittlenden gjøres mindre.
Vi kan starte med $\Delta t = 1$.
Hvis vi må regne ut veldig mange
trinn før vannstanden synker noe særlig, øker vi skrittlengden.


```python
# Setter inn tallverdier for A, B og g
A = 1
B = 0.001
g = 9.81

# Definerer funkionen f(t, x) = x
def f(t, x):
  return - (B/A) * np.sqrt(2 * g * np.maximum(x, 0))
```

For plotting gjenbruker vi koden fra før.

```python
# @title Plot med retningsdiagram og steg av metoden til Euler
N = 20


# Definerer t-verdiene
t = np.linspace(0, 600, num=N+1)


# Set the initial conditions and parameters
t0 = t[0]   # Første x-verdi
x0 = 1      # Innital x-verdi
tn = t[-1]  # Siste t-verdi
n = 10      # Antall steg for Eulers metode



# Anvend Eulers metode
t_values, x_values = euler_method(f, t0, x0, tn, n)

# Definerer x-verdiene for retningsvektorene 
x = np.linspace(x0, x_values[-1], num=N+1)

# Lager et firkantet gitter av de oppgitte x- og y-verdiene
T, X = np.meshgrid(t, x)

# Definerer f med hensyn på T og X
F = f(T, X)

# Definerer "steget" i x- og y-retning
dt = np.ones_like(T)
dx = F

# plt.quiver tar for seg å lage alle tangentpilene [dt, dx] i alle punktene som opptrer i gitteret vi lagde.
plt.quiver(T, X, dt, dx, scale=50, scale_units='width', color='black', angles='xy')
plt.plot(t_values, x_values, '-o', label=r'Tilnærmet løsning')
# plt.ylim((0, 50))
# plt.grid()
plt.xlabel('t')
plt.ylabel('x')
plt.legend(loc='upper left')
plt.title(r"Eulers metode for x′(t)=x(t) i " + f"{n} steg");
```

Vi gjenbruker koden fra før. Hvis vi vil lage mange slike figurer kan det vøre en hjelp å lage en funksjon `plot_retningsvektorer_og_euler_løsning(f, t, x0, antall_eulersteg, antall_piler)`  

```python
def plot_retningsvektorer_og_euler_løsning(f, t, x0, antall_eulersteg=10):    
    
    # Set the initial conditions and parameters
    t0 = t[0]   # Første x-verdi
    x0 = 1  # Iinital x-verdi
    tn = t[-1]  # Siste t-verdi
    n = antall_eulersteg   # Antall steg for Eulers metode
    
    
    # Anvend Eulers metode
    t_values, x_values = euler_method(f, t0, x0, tn, n)
    
    # Definerer x-verdiene for retningsvektorene 
    x = np.linspace(x0, x_values[-1], num=len(t))
    
    # Lager et firkantet gitter av de oppgitte x- og y-verdiene
    T, X = np.meshgrid(t, x)
    
    # Definerer f med hensyn på T og X
    F = f(T, X)
    
    # Definerer "steget" i x- og y-retning
    dt = np.ones_like(T)
    dx = F
    
    # plt.quiver tar for seg å lage alle tangentpilene [dt, dx] i alle punktene som opptrer i gitteret vi lagde.
    plt.quiver(T, X, dt, dx, color='black', angles='xy')
    plt.plot(t_values, x_values, '-o', label=r'Tilnærmet løsning')
    # plt.ylim((0, 50))
    # plt.grid()
    plt.xlabel('t')
    plt.ylabel('x')
    plt.legend(loc='upper left')
    plt.title(r"Eulers metode for x′(t)=x(t) i " + f"{len(t)} steg");   
```

## Om feil i Eulers metode

Det kan gjøres analyse av hvordan steglengden $\Delta t$ i Eulers metode påvirker nøyaktigheten av resultatet. For hvert steg som gjøres er feilen av orden $(\Delta t)^2$. Siden antall steg som gjøres for å komme fra initialverdien $t_0$ til sluttverdien er omvendt proporsjonalt med $\Delta t$ blir den globale feilen i Eulers metode av orden $\Delta t$.
