```python
n = 20
x = np.linspace(-2*np.pi, 2*np.pi, n)
y = np.sin(x)
dx = 4*np.pi / n
dy = np.gradient(y, dx)
```

```python
plt.plot(x,dy)
plt.plot(x, np.cos(x))
```
