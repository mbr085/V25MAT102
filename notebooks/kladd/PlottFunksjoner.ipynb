```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1, 100)
```

```python
def f(x):
    return np.exp(-x**2)
xx = np.linspace(-1,1, 100)
plt.plot(xx, f(xx))
plt.savefig('normal1.png')
```

```python
def f(x):
    return np.exp(-x**2)
xx = np.linspace(-3,3, 100)
plt.plot(xx, f(xx))
plt.savefig('normal3.png')
```

```python
def f(x):
    return 1-x**2
xx = np.linspace(-1,1, 100)
plt.plot(xx, f(xx))
plt.savefig('parabel.png')
```
