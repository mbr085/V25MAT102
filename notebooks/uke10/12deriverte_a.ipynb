# Derivasjon

## Vektorer og totalderiverte

En *vektorfunksjon* er en funksjon som tar en verdi i $\mathbb{R}^n$ og
gir en vektor i $\mathbb{R}^m$. En vektorfunksjon kan skrives som en
funksjon $\mathbf{F}:\mathbb{R}^n\to\mathbb{R}^m$ som tar en vektor
$\mathbf{x}\in\mathbb{R}^n$ og gir en vektor
$\mathbf{F}(\mathbf{x})\in\mathbb{R}^m$. En vektorfunksjon kan også
skrives som en liste av funksjoner, for eksempel
$$\mathbf{F}(\mathbf{x}) = \begin{bmatrix} F_1(\mathbf{x}) \\ F_2(\mathbf{x}) \\ \vdots \\ F_m(\mathbf{x}) \end{bmatrix}.$$

**Eksempel 1**. La $A$ være $2 \times 2$-matrisen
$$A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}.$$ Da er funksjonen
$\mathbf{F}\colon \mathbb{R}^2 \to \mathbb{R}^2$ gitt ved
$$\mathbf{F}(\mathbf{x}) = A \cdot \mathbf{x}= \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}
  \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} x_1 + 2x_2 \\ 3x_1
  + 4x_2 \end{bmatrix}.$$ en vektorfunksjon. Her skriver vi $\mathbf{x}$
for et punkt $\mathbf{x}= \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ i
$\mathbb{R}^2$. Siden et punkt i $\mathbb{R}^2$ er det samme som en
vektor, kan vi skrive $\mathbf{F}(\mathbf{x}) = A \cdot \mathbf{x}$. Det
er vanskelig å tegne grafen til denne funksjonen fordi den lever i fire
dimensjoner. Det vi kan gjøre er å velge noen vektorer og se hvor de
blir avbildet. For eksempel er
$$\mathbf{F}\left(\begin{bmatrix} 1 \\ 0 \end{bmatrix}\right) = \begin{bmatrix} 1 \\ 3 \end{bmatrix}.$$

**Eksempel 2**. Vi fortsetter eksempel
<a href="#lineærfunksjon" data-reference-type="ref"
data-reference="lineærfunksjon">1</a> og lar
$\mathbf{b} = \begin{bmatrix}b_1\\b_2\end{bmatrix}$ være et punkt i
$\mathbb{R}^2$. Da kan vi konstruere en vektorfunsjon
$\mathbf{G}\colon \mathbb{R}^2 \to \mathbb{R}^2$ gitt ved
$$\mathbf{G}(\mathbf{x}) = A \cdot \mathbf{x}+ \mathbf{b} = \begin{bmatrix} x_1 + 2x_2 \\ 3x_1 + 4x_2
    \end{bmatrix} + \begin{bmatrix} b_1 \\ b_2 \end{bmatrix} = \begin{bmatrix}
  x_1 + 2x_2 + b_1 \\ 3x_1 + 4x_2 + b_2 \end{bmatrix}.$$

**Eksempel 3**. Vanlige funksjoner $f \colon \mathbb{R}\to \mathbb{R}$
kan også oppfattes som vektorfunksjoner. Også funksjoner på formen
$f \colon \mathbb{R}^2 \to \mathbb{R}$ er vektorfunksjoner. Vi har
allerede sett eksempler på slike funksjoner og deres grafer.

**Definisjon 4**. Funksjonen $\mathbf{F}$ er *deriverbar* i punktet
$\mathbf{x}_0$ hvis det finnes en $m \times n$-matrise
$\mathbf{F}'(\mathbf{x}_0)$ slik at
$$\lim_{\mathbf{h}\to 0} \frac{\mathbf{F}(\mathbf{x}_0+\mathbf{h}) - \mathbf{F}(\mathbf{x}_0) -
\mathbf{F}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|} = 0.$$ Her tas
grensen over $\mathbf{h} \in \mathbb{R}^n$ og $| \mathbf{h}|$ er lengden
til vektoren $\mathbf{h}$.

Funksjonen $\mathbf{F}$ er *derivervar* hvis den er deriverbar i alle
punktene i $\mathbb{R}^n$.

Matrisen $\mathbf{F}(\mathbf{x}_0)'$ kalles den deriverte av
$\mathbf{F}$ i punktet $\mathbf{x}_0$. Et annet navn for denne matrisen
er *Jacobimatrisen* til $\mathbf{F}$ i punktet $\mathbf{x}_0$.

Vi går ikke inn på detalj om grenseverdier i flere variabler. Alt vi
trenger å vite er at de har samme regneregler som grenseverdier i en
variabel og at grenseverdier kan beregnes komponentvis. Dette medfører
at de generelle derivasjonsreglene for funksjoner av en variabel også
gjelder for vektorfunksjoner. En annen konsekvens er at
$$\mathbf{F}'(\mathbf{x}) = \begin{bmatrix} F'_1(\mathbf{x}) \\ F'_2(\mathbf{x})
\\ \vdots \\ F'_m(\mathbf{x}) \end{bmatrix},$$ hvor matrisen på høyre
side er en $m \times n$-matrise fremkommet ved å stable $m$ matriser av
formen $(1 \times n)$ på hverandre. Det vil si $m$ radvektorer, hvor
hver radvektor er den deriverte av en funksjon $F_i$.

**Eksempel 5**. La $\mathbf{G}$ være vektorfunksjonen fra eksempel
<a href="#affinfunksjon" data-reference-type="ref"
data-reference="affinfunksjon">2</a>. Da skal
$\mathbf{G}'(\mathbf{x}_0)$ være en matrise slik at
$$\lim_{\mathbf{h}\to 0} \frac{\mathbf{G}(\mathbf{x}_0+\mathbf{h}) - \mathbf{G}(\mathbf{x}_0) -
  \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|} = 0.$$
Regner vi ut dette får vi $$\begin{aligned}
    &
    \lim_{\mathbf{h}\to 0} \frac{\mathbf{G}(\mathbf{x}_0+\mathbf{h}) -
    \mathbf{G}(\mathbf{x}_0) -
    \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|} \\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{\left(A \cdot (\mathbf{x}_0+\mathbf{h})  +
    \mathbf{b}\right) - \left(A \cdot \mathbf{x}_0  - \mathbf{b}\right) -
    \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|}\\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{A \cdot (\mathbf{x}_0+\mathbf{h})  +
    A \cdot \mathbf{x}_0  - 
    \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|} \\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{A \cdot \mathbf{h}  - 
    \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|}.
  
\end{aligned}$$ Vi ser at $\mathbf{G}'(\mathbf{x}_0) = A$ er den
deriverte til $\mathbf{G}$ i punktet $\mathbf{x}_0$.

**Definisjon 6**. *Lineærtilnærmingen* til en vektorfunksjon
$\mathbf{F}$ som er deriverbar i punktet $\mathbf{x}_0$ er gitt ved
$$L(\mathbf{x}) = \mathbf{F}(\mathbf{x}_0) + \mathbf{F}'(\mathbf{x}_0) \cdot (\mathbf{x} - \mathbf{x}_0).$$

**Eksempel 7**. La igjen $\mathbf{G}$ være vektorfunksjonen fra eksempel
<a href="#affinfunksjon" data-reference-type="ref"
data-reference="affinfunksjon">2</a>. Lineærtilnærmingen til
$\mathbf{G}$ i punktet $\mathbf{x}_0$ er gitt ved
$$L(\mathbf{x}) = \mathbf{G}(\mathbf{x}_0) + \mathbf{G}'(\mathbf{x}_0) \cdot (\mathbf{x}- \mathbf{x}_0) = A \cdot \mathbf{x}_0 + \mathbf{b} + A \cdot (\mathbf{x}- \mathbf{x}_0).$$
Siden
$A \cdot \mathbf{x}_0 + \mathbf{b} + A \cdot (\mathbf{x}- \mathbf{x}_0) = A \cdot \mathbf{x}_0 + A
  \cdot \mathbf{x}- A \cdot \mathbf{x}_0 + \mathbf{b} = A \cdot \mathbf{x}+ \mathbf{b}$,
ser vi at $L(\mathbf{x}) = \mathbf{G}(\mathbf{x})$.

**Eksempel 8**. La $f \colon \mathbb{R}\to \mathbb{R}$ være en vanlig
funksjon av en variabel. Betingelsen for at $f$ er deriverbar i $x_0$
kan skrives:
$$\lim_{h\to 0} \frac{f(x_0+h) - f(x_0) - f'(x_0) \cdot h}{h} =
  \lim_{h\to 0} \frac{f(x_0+h) - f(x_0)}{h} - f'(x_0) = 0.$$ Her står
altså $$f'(x_0) = \lim_{h\to 0} \frac{f(x_0+h) - f(x_0)}{h}.$$ Dette er
den vanlige definisjonen av den deriverte av en funksjon av en variabel.

Lineærtilnærmingen til $f$ i punktet $x_0$ er gitt ved
$$L(x) = f(x_0) + f'(x_0) \cdot (x - x_0).$$ Grafen til $L(x)$ er
tangenten i punktet $x_0$ til grafen til $f$.

**Eksempel 9**. La $\mathbf{y} \colon \mathbb{R}\to \mathbb{R}^2$ være
en kurve i planet. Betingelsen for at $\mathbf{y}$ er deriverbar i $x_0$
kan skrives:
$$\lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0) - \mathbf{y}'(x_0) \cdot h}{h} =
  \lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0)}{h} - \mathbf{y}'(x_0) = 0.$$
Her står altså
$$\mathbf{y}'(x_0) = \lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0)}{h}.$$
Som bemerket over kan denne grensen beregens komponentvis slik at hvis
$$\mathbf{y}(t) =
  \begin{bmatrix} y_1(t) \\ y_2(t) \end{bmatrix},$$ så er
$$\mathbf{y}'(t) =
  \begin{bmatrix} y_1'(t) \\ y_2'(t) \end{bmatrix}.$$ Lineærtilnærmingen
til $\mathbf{y}$ i punktet $t_0$ er gitt ved
$$L(t) = \mathbf{y}(t_0) + \mathbf{y}'(t_0) \cdot (t - t_0).$$ Dette er
en parametrisering av tangenten til kurven i punktet $\mathbf{y}(t_0)$.

## Gradienten og retningsderiverte

**Definisjon 10**. *Den retningsderiverte* av en flervariabel funksjon
$f \colon \mathbb{R}^n \to
  \mathbb{R}$ i punktet $\mathbf{x}_0$ i en retning
$\vec{\mathbf{v}} \in \mathbb{R}^n$ er gitt ved
$$D_{\vec{\mathbf{v}}} f(\mathbf{x}_0) = \lim_{t\to 0} \frac{f(\mathbf{x}_0 + t\vec{\mathbf{v}}) - f(\mathbf{x}_0)}{t}.$$

Her legger vi et punkt $\mathbf{x}_0$ og en vektor $t \vec{\mathbf{v}}$
til punktet $\mathbf{x}_0$ og Gitt $\mathbf{x_0} \in \mathbb{R}^n$ og en
vektor $\vec{\mathbf{v}} \in \mathbb{R}^n$, er funksjonen
$$g(t) = \mathbf{x_0} + t\vec{\mathbf{v}}$$ en parametrisering av den
rette linjen gjennom $\mathbf{x_0}$ i retning $\vec{\mathbf{v}}$.
Funksjonen $f \circ g$ er en funksjon av en variabel, gitt ved
$(f \circ g)(t) = f(g(t))$. Den deriverte av denne kompositten i punktet
$t=0$ er retningsderiverte av $f$ i punktet $\mathbf{x_0}$ i retning
$\vec{\mathbf{v}}$. Det vil si at
$D_{\vec{\mathbf{v}}} f(\mathbf{x_0}) = (f \circ g)'(0)$.

Hvis $D_{\vec{\mathbf{v}}} f(\mathbf{x_0}) > 0$ da vokser funksjonen $f$
når vi beveger oss vekk fra punketet $\mathbf{x}_0$ i retning
$\vec{\mathbf{v}}$. Hvis $D_{\vec{\mathbf{v}}} f(\mathbf{x_0}) < 0$ da
avtar funksjonen $f$ når vi beveger oss vekk fra punktet $\mathbf{x}_0$
i retning $\vec{\mathbf{v}}$.

**Oppgave 1**. Forklar hvorfor
$D_{\vec{\mathbf{v}}} f(\mathbf{x}_0) = f'(\mathbf{x}_0)\cdot \vec{\mathbf{v}}$.

Husk at enhetsvektoren $\vec{\mathbf{e}}_i$ er en vektor med lengde 1 i
retning $i$-aksen. Det vil si at
$\vec{\mathbf{e}}_i = \begin{bmatrix} 0 \\ \vdots \\ 1 \\ \vdots \\ 0 \end{bmatrix}$,
der 1 står på $i$-te plass.

**Definisjon 11**. Den retningsderiverte av $f$ i punktet $\mathbf{x}_0$
i retning $\vec{\mathbf{e}}_i$ kalles den $i$-te partiellderiverte av
$f$ i punktet $\mathbf{x}_0$ og skrives
$$\frac{\partial f}{\partial x_i}(\mathbf{x}_0) = D_{\vec{\mathbf{e}}_i} f(\mathbf{x}_0).$$

Fra oppgave
<a href="#retningsderivertderivert" data-reference-type="ref"
data-reference="retningsderivertderivert">1</a> har vi at
$$\frac{\partial f}{\partial x_i}(\mathbf{x}_0) = f'(\mathbf{x}_0)\cdot \vec{\mathbf{e}}_i.$$
Av dette kan vi lese av at
$$f'(\mathbf{x}_0) = \begin{bmatrix} \frac{\partial f}{\partial x_1}(\mathbf{x}_0) & \dots & \frac{\partial f}{\partial x_n}(\mathbf{x}_0) \end{bmatrix}.$$

For en vektorfunksjon $\mathbf{F}\colon \mathbb{R}^n \to \mathbb{R}^m$
på formen
$$\mathbf{F}(\mathbf{x}) = \begin{bmatrix} F_1(\mathbf{x}) \\ F_2(\mathbf{x}) \\ \vdots \\ F_m(\mathbf{x}) \end{bmatrix},$$
har vi
$$\mathbf{F}'(\mathbf{x}_0) = \begin{bmatrix} F'_1(\mathbf{x}_0) \\ F'_2(\mathbf{x}_0) \\ \vdots \\ F'_m(\mathbf{x}_0) \end{bmatrix}
= 
\begin{bmatrix}
  \frac{\partial F_1}{\partial x_1}
  & \dots &
  \frac{\partial F_1}{\partial x_n}
  \\
  \frac{\partial F_2}{\partial x_1}
  & \dots &
  \frac{\partial F_2}{\partial x_n}
  \\
  \vdots &&  \vdots \\
  \frac{\partial F_m}{\partial x_1}
         & \dots&
  \frac{\partial F_m}{\partial x_n}
  \end{bmatrix}.$$

**Oppgave 2**. Per definisjon er den retningsderiverte av en
vektorfunksjon den deriverte til en funksjon av en variabel. Beskriv en
funksjon $h \colon \mathbb{R}\to \mathbb{R}$ slik at
$\frac{\partial f}{\partial x_i}(\mathbf{x}_0) = h'(0)$.

**Definisjon 12**. *Gradienten* til en vektorfunksjon
$f \colon \mathbb{R}^n \to \mathbb{R}$ i et punkt
$\mathbf{x}_0 \in \mathbb{R}^n$ er vektoren
$$\nabla f(\mathbf{x}_0) = \begin{bmatrix}\frac{\partial f}{\partial x_1}(\mathbf{x}_0) \\ \vdots \\ \frac{\partial f}{\partial x_n}(\mathbf{x}_0)\end{bmatrix}$$

*Bemerkning 13*. Gradienten $\nabla f(\mathbf{x}_0)$ er en
kolonnevektor, og $f'(\mathbf{x}_0)$ er radvektoren som fås ved å legge
denne kolonnevktoren ned som vi gjorde da vi definerte
matrisemultiplikasjon.

Teoremet sammenfatter utsagent i Oppgave
<a href="#retningsderivertderivert" data-reference-type="ref"
data-reference="retningsderivertderivert">1</a> og sammenhengen mellom
prikkprodukt og vinkler mellom vektorer. Det gir en ny måte å se på
gradienten.

**Teorem 14**. *Hvis $f\colon \mathbb{R}^n \to \mathbb{R}$ er en
skalarfunksjon og $\mathbf{x}_0$ er et punkt slik at $f'(\mathbf{x}_0)$
er definert, så er
$$D_{\vec{\mathbf{v}}} f(\mathbf{x}_0) = f'(\mathbf{x}_0) \cdot \vec{\mathbf{v}}= \nabla f(\mathbf{x}_0) \cdot
  \vec{\mathbf{v}} = |\nabla f(\mathbf{x}_0)| \cos \theta,$$ for alle
vektorer $\vec{\mathbf{v}}$ med lengde $1$, der $\theta$ er vinkelen
mellom $\vec{\mathbf{v}}$ og $\nabla f(\mathbf{x}_0)$. Derfor er
$\nabla f(\mathbf{x}_0)$ retningen der $f$ vokser raskest i punktet
$\mathbf{x}_0$.*

## Gradienter og nivåmengder

**Definisjon 15**. En *nivåkurve* til en funksjon
$f \colon \mathbb{R}^n \to \mathbb{R}$ er en kurve
$\mathbf{r} \colon \mathbb{R}\to \mathbb{R}^n$ slik at
$f(\mathbf{r}(t)) = c$ for en konstant $c$.

**Teorem 16**. *Gitt $f \colon \mathbb{R}^n \to \mathbb{R}$ og en
nivåkurve $\mathbf{r} \colon \mathbb{R}\to \mathbb{R}^n$ til $f$ og et
tall $t$ slik at de deriverte $\mathbf{r}'(t)$ og $f'(\mathbf{r}(t))$
begge er definert. Da er
$$\nabla f(\mathbf{r}(t)) \cdot \mathbf{r}'(t) = 0.$$*

Dette er en direkte konsekvens av Oppgave
<a href="#retningsderivertderivert" data-reference-type="ref"
data-reference="retningsderivertderivert">1</a> og kjerneregelen:
$$0 = (f \circ \mathbf{r})'(t) =
  f'(\mathbf{r}(t)) \cdot \mathbf{r}'(t) =
  \nabla f(\mathbf{r}(t)) \cdot \mathbf{r}'(t).$$
