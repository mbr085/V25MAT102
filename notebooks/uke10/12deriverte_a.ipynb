## Vektorer og totalderiverte

En *vektorfunksjon* er en funksjon som tar en verdi i $\mathbb{R}^n$ og
gir en vektor i $\mathbb{R}^m$. En vektorfunksjon kan skrives som en
funksjon $\mathbf{F}:\mathbb{R}^n\to\mathbb{R}^m$ som tar en vektor
$\mathbf{x}\in\mathbb{R}^n$ og gir en vektor
$\mathbf{F}(\mathbf{x})\in\mathbb{R}^m$. En vektorfunksjon kan også
skrives som en liste av funksjoner, for eksempel
$$\mathbf{F}(\mathbf{x}) = \begin{bmatrix} F_1(\mathbf{x}) \\ F_2(\mathbf{x}) \\ \vdots \\ F_m(\mathbf{x}) \end{bmatrix}.$$
I teksten under kommer jeg ofte til å skrive "funksjon" når jeg mener
"vektorfunksjon".

**Eksempel 1**. La $A$ være $2 \times 2$-matrisen
$$A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}.$$ Da er funksjonen
$\mathbf{F}\colon \mathbb{R}^2 \to \mathbb{R}^2$ gitt ved
$$\mathbf{F}(\mathbf{x}) = A \cdot \mathbf{x}= \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}
  \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} x_1 + 2x_2 \\ 3x_1
  + 4x_2 \end{bmatrix}.$$ en vektorfunksjon.

Her skriver vi $\mathbf{x}$ for et punkt
$\mathbf{x}= \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ i $\mathbb{R}^2$.
Siden et punkt i $\mathbb{R}^2$ er det samme som en vektor, kan vi
skrive $\mathbf{F}(\mathbf{x}) = A \cdot \mathbf{x}$. Det er vanskelig å
tegne grafen til denne funksjonen fordi den lever i fire dimensjoner.
Det vi kan gjøre er å velge noen vektorer og se hvorhen de blir
avbildet. For eksempel er
$$\mathbf{F}\left(\begin{bmatrix} 1 \\ 0 \end{bmatrix}\right) = \begin{bmatrix} 1 \\ 3 \end{bmatrix}.$$

```python
# @title Import
import matplotlib.pyplot as plt
import numpy as np
```

```python
# @title Plotter
# Definer matrisen A
A = np.array([[1, 2], [3, 4]])

# Definer vektorene
v1 = np.array([1, 0])
v2 = np.array([0, 1])
v3 = np.array([1, 1])

# Anvend transformasjonen
Av1 = A @ v1  # Av1 er resultatet av å multiplisere A med v1
Av2 = A @ v2  # Av2 er resultatet av å multiplisere A med v2
Av3 = A @ v3  # Av3 er resultatet av å multiplisere A med v3

# Lag plottet
plt.figure(figsize=(8, 8))  # Sett størrelsen på figuren

# Plott de originale vektorene
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')  # Plott v1 i blått
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')  # Plott v2 i rødt
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='green', label='v3')  # Plott v3 i grønt

# Plott de transformerte vektorene (bildene)
plt.quiver(0, 0, Av1[0], Av1[1], angles='xy', scale_units='xy', scale=1, color='blue', linestyle='dashed')  # Plott Av1 med stiplet linje
plt.quiver(0, 0, Av2[0], Av2[1], angles='xy', scale_units='xy', scale=1, color='red', linestyle='dashed')  # Plott Av2 med stiplet linje
plt.quiver(0, 0, Av3[0], Av3[1], angles='xy', scale_units='xy', scale=1, color='green', linestyle='dashed')  # Plott Av3 med stiplet linje

# Legg til navn på de originale vektorene
plt.annotate(r"$\mathbf{x}$", v1 + 0.1, color='blue')  # Legg til navn på v1, litt forskjøvet for bedre synlighet
plt.annotate(r"$\mathbf{y}$", v2 + 0.1, color='red')  # Legg til navn på v2, litt forskjøvet for bedre synlighet
plt.annotate(r"$\mathbf{z}$", v3 + 0.1, color='green')  # Legg til navn på v3, litt forskjøvet for bedre synlighet

# Legg til navn på de transformerte vektorene
plt.annotate(r"$\mathbf{F}(\mathbf{x})$", Av1 + [-0.1, 0.1], color='blue')  # Legg til navn på Av1, litt forskjøvet
plt.annotate(r"$\mathbf{F}(\mathbf{y})$", Av2 + 0.1, color='red')  # Legg til navn på Av2, litt forskjøvet
plt.annotate(r"$\mathbf{F}(\mathbf{z})$", Av3 + 0.1, color='green')  # Legg til navn på Av3, litt forskjøvet

matrix_text = u"⎡1 2⎤\n⎣3 4⎦"
plt.text(5, 8, matrix_text, fontsize=12)
plt.text(4.4, 8.2, 'A =', fontsize=12)

# Legg til aksenavn og tittel
plt.xlabel("$x_1$")  # Navn på x-aksen
plt.ylabel("$x_2$")  # Navn på y-aksen
plt.xlim(-1, 7)  # Sett grenser for x-aksen
plt.ylim(-1, 9)  # Sett grenser for y-aksen
plt.title("Transformasjon av vektorer ved matrise A")  # Tittel på plottet
plt.grid(True)  # Vis rutenett
#plt.legend()  # Vis forklaring (kan aktiveres om ønskelig)

# Vis plottet
plt.show()
```

**Eksempel 3**. Vanlige funksjoner $f \colon \mathbb{R}\to \mathbb{R}$
kan også oppfattes som vektorfunksjoner. Også flervariabel funksjoner på
formen $f \colon \mathbb{R}^2 \to \mathbb{R}$ er vektorfunksjoner. Vi
har allerede sett eksempler på slike funksjoner og deres grafer.

**Definisjon 4**. Funksjonen $\mathbf{F}$ er *deriverbar* i punktet
$\mathbf{x}_0$ hvis det finnes en $m \times n$-matrise
$\mathbf{F}'(\mathbf{x}_0)$ slik at
$$\lim_{\mathbf{h}\to 0} \frac{\mathbf{F}(\mathbf{x}_0+\mathbf{h}) - \mathbf{F}(\mathbf{x}_0) -
  \mathbf{F}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|} = 0.$$ Her
tas grensen over $\mathbf{h} \in \mathbb{R}^n$ og $| \mathbf{h}|$ er
lengden til vektoren $\mathbf{h}$.

Funksjonen $\mathbf{F}$ er *derivervar* hvis den er deriverbar i alle
punktene i $\mathbb{R}^n$.

Matrisen $\mathbf{F}(\mathbf{x}_0)'$ kalles den deriverte av
$\mathbf{F}$ i punktet $\mathbf{x}_0$. Et annet navn for denne matrisen
er *Jacobimatrisen* til $\mathbf{F}$ i punktet $\mathbf{x}_0$.

Vi går ikke inn på detalj om grenseverdier i flere variabler. Alt vi
trenger å vite er at de har samme regneregler som grenseverdier i en
variabel og at grenseverdier kan beregnes komponentvis. En konsekvens er
at de generelle derivasjonsreglene for funksjoner av en variabel også
gjelder for vektorfunksjoner. En annen konsekvens er at
$$\mathbf{F}'(\mathbf{x}) = \begin{bmatrix} F'_1(\mathbf{x}) \\ F'_2(\mathbf{x})
\\ \vdots \\ F'_m(\mathbf{x}) \end{bmatrix},$$ hvor matrisen på høyre
side er en $m \times n$-matrise fremkommet ved å stable $m$ matriser av
formen $(1 \times n)$ på hverandre. Det vil si $m$ radvektorer, hvor
hver radvektor er den deriverte av en funksjon $F_i$.

**Eksempel 5**. La $\mathbf{G}$ være vektorfunksjonen fra eksempel
<a href="#affinfunksjon" data-reference-type="ref"
data-reference="affinfunksjon">2</a>. Da skal
$\mathbf{G}'(\mathbf{x}_0)$ være en matrise slik at
$$\lim_{\mathbf{h}\to 0} \frac{\mathbf{G}(\mathbf{x}_0+\mathbf{h}) - \mathbf{G}(\mathbf{x}_0) -
  \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|} = 0.$$
Regner vi på dette får vi:
$$\begin{align*}
    &
    \lim_{\mathbf{h}\to 0} \frac{\mathbf{G}(\mathbf{x}_0+\mathbf{h}) -
    \mathbf{G}(\mathbf{x}_0) -
    \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|} \\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{\left(A \cdot (\mathbf{x}_0+\mathbf{h})  +
    \mathbf{b}\right) - \left(A \cdot \mathbf{x}_0  - \mathbf{b}\right) -
    \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|}\\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{A \cdot (\mathbf{x}_0+\mathbf{h})  +
    A \cdot \mathbf{x}_0  -
    \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|} \\
    =
    &
    \lim_{\mathbf{h}\to 0} \frac{A \cdot \mathbf{h}  -
    \mathbf{G}'(\mathbf{x}_0) \cdot \mathbf{h}}{|\mathbf{h}|}.
\end{align*}$$
Vi ser at $\mathbf{G}'(\mathbf{x}_0) = A$ er den
deriverte til $\mathbf{G}$ i punktet $\mathbf{x}_0$.

**Definisjon 6**. *Lineærtilnærmingen* til en vektorfunksjon
$\mathbf{F}$ som er deriverbar i punktet $\mathbf{x}_0$ er gitt ved
$$L(\mathbf{x}) = \mathbf{F}(\mathbf{x}_0) + \mathbf{F}'(\mathbf{x}_0) \cdot (\mathbf{x} - \mathbf{x}_0).$$

**Eksempel 7**. La igjen $\mathbf{G}$ være vektorfunksjonen fra eksempel
<a href="#affinfunksjon" data-reference-type="ref"
data-reference="affinfunksjon">2</a>. Lineærtilnærmingen til
$\mathbf{G}$ i punktet $\mathbf{x}_0$ er gitt ved
$$L(\mathbf{x}) = \mathbf{G}(\mathbf{x}_0) + \mathbf{G}'(\mathbf{x}_0) \cdot (\mathbf{x}- \mathbf{x}_0) = A \cdot \mathbf{x}_0 + \mathbf{b} + A \cdot (\mathbf{x}- \mathbf{x}_0).$$
Siden
$A \cdot \mathbf{x}_0 + \mathbf{b} + A \cdot (\mathbf{x}- \mathbf{x}_0) = A \cdot \mathbf{x}_0 + A
  \cdot \mathbf{x}- A \cdot \mathbf{x}_0 + \mathbf{b} = A \cdot \mathbf{x}+ \mathbf{b}$,
ser vi at $L(\mathbf{x}) = \mathbf{G}(\mathbf{x})$.

**Eksempel 8**. La $f \colon \mathbb{R}\to \mathbb{R}$ være en vanlig
funksjon. Betingelsen for at $f$ er deriverbar i $x_0$ kan skrives:
$$\lim_{h\to 0} \frac{f(x_0+h) - f(x_0) - f'(x_0) \cdot h}{h} =
  \lim_{h\to 0} \frac{f(x_0+h) - f(x_0)}{h} - f'(x_0) = 0.$$ Her står
altså $$f'(x_0) = \lim_{h\to 0} \frac{f(x_0+h) - f(x_0)}{h}.$$ Dette er
den vanlige definisjonen av den deriverte av en funksjon.

Lineærtilnærmingen til $f$ i punktet $x_0$ er gitt ved
$$L(x) = f(x_0) + f'(x_0) \cdot (x - x_0).$$ Grafen til $L(x)$ er
tangenten i punktet $x_0$ til grafen til $f$.

```python
# @title Tangent
# Definer funksjonen
def f(x):
  return (x - 1) * (x - 1) * (x - 2)  # Funksjonen f(x) = (x-1)²(x-2)

# Definer den deriverte av funksjonen
def df(x):
  return 2 * (x - 1) * (x - 2) + (x - 1) * (x - 1)  # Den deriverte av f(x)

# Definer intervallet
x = np.linspace(0, 3, 400)  # Lag 400 jevnt fordelte punkter mellom 0 og 3

# Beregn funksjonsverdiene
y = f(x)  # Beregn f(x) for alle punktene i x

# Punkt for tangenten
x0 = 2  # x-verdien der vi vil tegne tangenten
y0 = f(x0)  # y-verdien til tangeringspunktet
m = df(x0)  # Stigningstallet til tangenten

# Beregn tangentlinjen
x_tan = np.linspace(1., 3., 100)  # Intervall for tangentlinjen
y_tan = m * (x_tan - x0) + y0  # Beregn y-verdiene til tangentlinjen

# Lag plottet
plt.plot(x, y, label=r'$f(x) = (x-1)^2(x-2)$')  # Plott funksjonen
plt.plot(x_tan, y_tan, label='Tangent ved x=2')  # Plott tangentlinjen
plt.scatter(x0, y0, color='red', label='Tangeringspunkt')  # Marker tangeringspunktet

# Tilpass plottet
plt.xlabel('x')  # Navn på x-aksen
plt.ylabel('f(x)')  # Navn på y-aksen
plt.title('Plott av f(x) og dens tangent ved x=2')  # Tittel på plottet
plt.legend()  # Vis forklaring
plt.grid(True)  # Vis rutenett

# Vis plottet
plt.show()
```

**Eksempel 9**. La $\mathbf{y} \colon \mathbb{R}\to \mathbb{R}^2$ være
en kurve i planet. Betingelsen for at $\mathbf{y}$ er deriverbar i $x_0$
kan skrives:
$$\lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0) - \mathbf{y}'(x_0) \cdot h}{h} =
  \lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0)}{h} - \mathbf{y}'(x_0) = 0.$$
Her står altså
$$\mathbf{y}'(x_0) = \lim_{h\to 0} \frac{\mathbf{y}(x_0+h) - \mathbf{y}(x_0)}{h}.$$
Som bemerket over kan denne grensen beregens komponentvis slik at hvis
$$\mathbf{y}(t) =
  \begin{bmatrix} y_1(t) \\ y_2(t) \end{bmatrix},$$ så er
$$\mathbf{y}'(t) =
  \begin{bmatrix} y_1'(t) \\ y_2'(t) \end{bmatrix}.$$ Lineærtilnærmingen
til $\mathbf{y}$ i punktet $t_0$ er gitt ved
$$L(t) = \mathbf{y}(t_0) + \mathbf{y}'(t_0) \cdot (t - t_0).$$ Dette er
en parametrisering av tangenten til kurven i punktet $\mathbf{y}(t_0)$.

```python
# @title Tangent
# Definer funksjonen
def f(x):
  return (x - 1) * (x - 1) * (x - 2)  # Funksjonen f(x) = (x-1)²(x-2)

# Definer den deriverte av funksjonen
def df(x):
  return 2 * (x - 1) * (x - 2) + (x - 1) * (x - 1)  # Den deriverte av f(x)

# Definer intervallet
x = np.linspace(0, 3, 400)  # Lag 400 jevnt fordelte punkter mellom 0 og 3

# Beregn funksjonsverdiene
y = f(x)  # Beregn f(x) for alle punktene i x

# Punkt for tangenten
x0 = 2  # x-verdien der vi vil tegne tangenten
y0 = f(x0)  # y-verdien til tangeringspunktet
m = df(x0)  # Stigningstallet til tangenten

# Beregn tangentlinjen
x_tan = np.linspace(1., 3, 100)  # Intervall for tangentlinjen
y_tan = m * (x_tan - x0) + y0  # Beregn y-verdiene til tangentlinjen

# Lag plottet
plt.plot(x, y, label=r'$y(t) = (t, \ (t-1)^2(t-2))$')  # Plott funksjonen
plt.plot(x_tan, y_tan, label='Tangent ved x=2')  # Plott tangentlinjen
plt.scatter(x0, y0, color='red', label='Tangeringspunkt')  # Marker tangeringspunktet

# Tilpass plottet
plt.xlabel('x')  # Navn på x-aksen
plt.ylabel('f(x)')  # Navn på y-aksen
plt.title('Plott av y(t) og dens tangent ved t=2')  # Tittel på plottet
plt.legend()  # Vis forklaring
plt.grid(True)  # Vis rutenett

# Vis plottet
plt.show()
```

```python
# @title Tangent
def yy(t):
    return np.vstack([2*np.cos(t), np.sin(t)])

def dyy(t):
    return np.vstack([-2*np.sin(t), np.cos(t)])

# Definer intervallet
t = np.linspace(0, 2*np.pi, 400)  # Lag 400 jevnt fordelte punkter mellom 0 og 3

# Beregn funksjonsverdiene
y = yy(t)  # Beregn f(x) for alle punktene i x

# Punkt for tangenten
t0 = np.pi / 4

# Beregn tangentlinjen
t_tan = np.linspace(-1, +1, 100)  # Intervall for tangentlinjen
tan = yy(t0) + t_tan * dyy(t0)
# Lag plottet
plt.plot(*yy(t), label=r'$y(t) = (2\cos(t), \sin(t))$')  # Plott funksjonen
plt.plot(*tan, label='Tangent ved x=2')  # Plott tangentlinjen
plt.scatter(*yy(t0), color='red', label='Tangeringspunkt')  # Marker tangeringspunktet

# Tilpass plottet
plt.xlabel('x')  # Navn på x-aksen
plt.ylabel('f(x)')  # Navn på y-aksen
plt.title('Plott av y(t) og dens tangent ved t=2')  # Tittel på plottet
plt.legend()  # Vis forklaring
plt.grid(True)  # Vis rutenett
```

## Gradienten og retningsderiverte

**Definisjon 10**. *Den retningsderiverte* av en flervariabel funksjon
$f \colon \mathbb{R}^n \to
  \mathbb{R}$ i punktet $\mathbf{x}_0$ i en retning
$\vec{\mathbf{v}} \in \mathbb{R}^n$ er gitt ved
$$D_{\vec{\mathbf{v}}} f(\mathbf{x}_0) = \lim_{t\to 0} \frac{f(\mathbf{x}_0 + t\vec{\mathbf{v}}) - f(\mathbf{x}_0)}{t}.$$

Her legger vi et punkt $\mathbf{x}_0$ og en vektor $t \vec{\mathbf{v}}$
til punktet $\mathbf{x}_0$ og Gitt $\mathbf{x_0} \in \mathbb{R}^n$ og en
vektor $\vec{\mathbf{v}} \in \mathbb{R}^n$, er funksjonen
$$g(t) = \mathbf{x_0} + t\vec{\mathbf{v}}$$ en parametrisering av den
rette linjen gjennom $\mathbf{x_0}$ i retning $\vec{\mathbf{v}}$.
Funksjonen $f \circ g$ er en funksjon av en variabel, gitt ved
$(f \circ g)(t) = f(g(t))$. Den deriverte av denne sammensatte
funksjonen i punktet $t=0$ er retningsderiverte av $f$ i punktet
$\mathbf{x_0}$ i retning $\vec{\mathbf{v}}$. Det vil si at
$D_{\vec{\mathbf{v}}} f(\mathbf{x_0}) = (f \circ g)'(0)$.

Hvis $D_{\vec{\mathbf{v}}} f(\mathbf{x_0}) > 0$ da vokser funksjonen $f$
når vi beveger oss vekk fra punketet $\mathbf{x}_0$ i retning
$\vec{\mathbf{v}}$. Hvis $D_{\vec{\mathbf{v}}} f(\mathbf{x_0}) < 0$ da
avtar funksjonen $f$ når vi beveger oss vekk fra punktet $\mathbf{x}_0$
i retning $\vec{\mathbf{v}}$.

**Eksempel 4**. La $f \colon \mathbb{R}^2 \to \mathbb{R}$ være gitt ved
$$f(\begin{bmatrix} x_1 \\ x_2 \end{bmatrix}) =
  2e^{-((x_1 + 1)^2 + x_2^2)}
  +
  e^{-((x_1 - 2)^2 + x_2^2)}
  .$$ Vi plotter denne funksjonen over rektanglet
$[-2.5,3.5] \times [-1.5, 1.5]$ der $-2.5 \le x_1 \le 3.5$ og
$-1.5 \le x_2 \le 1.5$.

Dessuten lar vi $\vec{\mathbf{v}} = \begin{bmatrix}1\\0\end{bmatrix}$ og $\mathbf{x}_0 = \begin{bmatrix}0\\0\end{bmatrix}$.
Vi skriver $g(t) = \mathbf{x_0} + t\vec{\mathbf{v}}$ og
$h(t) = \mathbf{x_0} + t\vec{\mathbf{v}} + f(g(t)) \cdot \vec{\mathbf{e}}_3$ for kurven på grafen til $f$ over kurven gitt ved $g$ som har
$h'(0) = \vec{\mathbf{v}} + (f \circ g)'(0) \cdot \vec{\mathbf{e}}_3 = \vec{\mathbf{v}} + D_{\vec{\mathbf{v}}} f(\mathbf{x_0}) \cdot \vec{\mathbf{e}}_3$.

Her er $\vec{\mathbf{e}}_3 = \begin{bmatrix}0\\0\\1\end{bmatrix}$, og tillater oss å la $\vec{\mathbf{v}} = \begin{bmatrix}1\\0\\0\end{bmatrix}$ og ${\mathbf{x}}_0 = \begin{bmatrix}0\\0\\0\end{bmatrix}$i definisjonen av $h$.

```python
# @title Definerer funksjoner

def numerisk_gradient(funksjon, punkt, epsilon=1e-4):
    gradient = np.zeros_like(punkt)
    for idx in range(punkt.shape[0]):
        v = np.zeros_like(punkt)
        v[idx] = 1
        gradient[idx] = numerisk_retningsderivert(funksjon, punkt, v, epsilon)
    return gradient

def numerisk_retningsderivert(funksjon, punkt, v, epsilon=1e-4):
    q = punkt + epsilon * v
    return (funksjon(q) - funksjon(punkt))/epsilon

def kopier_akse(fig, ax):
    subplot_spec = ax.get_subplotspec()

    ax2 = fig.add_subplot(subplot_spec.get_gridspec().nrows, subplot_spec.get_gridspec().ncols, subplot_spec.num1 + 1, projection='3d', label="quiver_layer1")
    ax2.patch.set_alpha(0)
    # Set the same limits and viewing angle for ax2

    ax2.grid(False)  # Remove grid lines
    ax2.xaxis.pane.fill = False  # Hide X pane
    ax2.yaxis.pane.fill = False  # Hide Y pane
    ax2.zaxis.pane.fill = False  # Hide Z pane
    ax2.set_xticks([])  # Remove X ticks
    ax2.set_yticks([])  # Remove Y ticks
    ax2.set_zticks([])  # Remove Z ticks

    # Set the aspect ratio to be equal
    ax2.set_xlim(ax.get_xlim())  # Same x-limits
    ax2.set_ylim(ax.get_ylim())  # Same y-limits
    ax2.set_zlim(ax.get_zlim())  # Same z-limits

    # Ensure the same view angle
    ax2.view_init(elev=ax.elev, azim=ax.azim)  # Copy elevation and azimuthal angles from ax1
    ax2.set_box_aspect(ax.get_box_aspect())
    return ax2

def f(x):
  x1, x2 = x
  return 2*np.exp(-((x1 + 1)**2 + x2**2)) + np.exp(-((x1 - 2)**2 + x2**2))
```

**Oppgave 1**. Forklar hvorfor
$D_{\vec{\mathbf{v}}} f(\mathbf{x}_0) = f'(\mathbf{x}_0)\cdot \vec{\mathbf{v}}$.

Husk at enhetsvektoren $\vec{\mathbf{e}}_i$ er en vektor med lengde 1 i
retning $i$-aksen. Det vil si at
$\vec{\mathbf{e}}_i = \begin{bmatrix} 0 \\ \vdots \\ 1 \\ \vdots \\ 0 \end{bmatrix}$,
der 1 står på $i$-te plass.

**Definisjon 11**. Den retningsderiverte av $f$ i punktet $\mathbf{x}_0$
i retning $\vec{\mathbf{e}}_i$ kalles den $i$-te partiellderiverte av
$f$ i punktet $\mathbf{x}_0$ og skrives
$$\frac{\partial f}{\partial x_i}(\mathbf{x}_0) = D_{\vec{\mathbf{e}}_i} f(\mathbf{x}_0).$$

Fra oppgave
<a href="#retningsderivertderivert" data-reference-type="ref"
data-reference="retningsderivertderivert">1</a> har vi at
$$\frac{\partial f}{\partial x_i}(\mathbf{x}_0) = f'(\mathbf{x}_0)\cdot \vec{\mathbf{e}}_i.$$
Gitt en radvektor $\vec{\mathbf{v}}= [v_1, \dots, v_n]$ er
$\vec{\mathbf{v}}= [\vec{\mathbf{v}}\cdot \vec{\mathbf{e}}_1, \dots, \vec{\mathbf{v}}\cdot \vec{\mathbf{e}}_n]$.
Derfor er
$$f'(\mathbf{x}_0) = \begin{bmatrix} \frac{\partial f}{\partial x_1}(\mathbf{x}_0) & \dots & \frac{\partial f}{\partial x_n}(\mathbf{x}_0) \end{bmatrix}.$$

For en vektorfunksjon $\mathbf{F}\colon \mathbb{R}^n \to \mathbb{R}^m$
på formen
$$\mathbf{F}(\mathbf{x}) = \begin{bmatrix} F_1(\mathbf{x}) \\ F_2(\mathbf{x}) \\ \vdots \\ F_m(\mathbf{x}) \end{bmatrix},$$
har vi
$$\mathbf{F}'(\mathbf{x}_0) = \begin{bmatrix} F'_1(\mathbf{x}_0) \\ F'_2(\mathbf{x}_0) \\ \vdots \\ F'_m(\mathbf{x}_0) \end{bmatrix}
=
\begin{bmatrix}
  \frac{\partial F_1}{\partial x_1}
  & \dots &
  \frac{\partial F_1}{\partial x_n}
  \\
  \frac{\partial F_2}{\partial x_1}
  & \dots &
  \frac{\partial F_2}{\partial x_n}
  \\
  \vdots &&  \vdots \\
  \frac{\partial F_m}{\partial x_1}
         & \dots&
  \frac{\partial F_m}{\partial x_n}
  \end{bmatrix}.$$

funksjonen $\mathbf{G}\circ \mathbf{F}\colon \mathbb{R}^n \to
  \mathbb{R}^k$ differensierbar i $\mathbf{x}_0$, og
$$(\mathbf{G}\circ \mathbf{F})'(\mathbf{x}_0) =  \mathbf{G}'(\mathbf{y}_0) \cdot  \mathbf{F}'(\mathbf{x}_0).$$*

**Eksempel 16**. Vi kan bruke denne versjonen av kjerneregelen til å se
at $\frac{d}{dx} x^2 =
  2x$. La $\mathbf{F}\colon \mathbb{R}\to \mathbb{R}^2$ være gitt ved
$\mathbf{F}(x) = (x, x)$ og $\mathbf{G}\colon
  \mathbb{R}^2 \to \mathbb{R}$ være gitt ved
$\mathbf{G}(x, y) = x \cdot y$. De deriverte av $\mathbf{F}$ og
$\mathbf{G}$ er gitt ved
$$\mathbf{F}'(x) = \begin{bmatrix} 1 \\ 1 \end{bmatrix} \quad \text{og} \quad \mathbf{G}'(x, y)
  = \begin{bmatrix} y & x \end{bmatrix}.$$ Da er
$\mathbf{G}\circ \mathbf{F}(x) = x^2$, og vi kan bruke kjerneregelen til
å regne ut den deriverte:
$$\begin{aligned}
    \frac{d}{dx} x^2 &= (\mathbf{G}\circ \mathbf{F})'(x) = \mathbf{G}'(x, x) \cdot \mathbf{F}'(x) \\
    &= \begin{bmatrix} x & x \end{bmatrix} \begin{bmatrix} 1 \\ 1 \end{bmatrix} = x + x = 2x.    
\end{aligned}$$

**Eksempel 17**. Vi kan bruke denne versjonen av kjerneregelen til
forklare produktregelen for derivasjon. La
$f \colon \mathbb{R}\to \mathbb{R}$ og
$g \colon \mathbb{R}\to \mathbb{R}$ være to funksjoner som er
differensierbare i $x_0$. La
$\mathbf{F}\colon \mathbb{R}\to \mathbb{R}^2$ være gitt ved
$\mathbf{F}(x) = (f(x), g(x))$ og $\mathbf{G}\colon
  \mathbb{R}^2 \to \mathbb{R}$ være gitt ved
$\mathbf{G}(x, y) = x \cdot y$. De deriverte av $\mathbf{F}$ og
$\mathbf{G}$ er gitt ved
$$\mathbf{F}'(x) = \begin{bmatrix} f'(x) \\ g'(x) \end{bmatrix} \quad \text{og} \quad \mathbf{G}'(x, y)
  = \begin{bmatrix} y & x \end{bmatrix}.$$ Da er
$\mathbf{G}\circ \mathbf{F}(x) = x^2$, og vi kan bruke kjerneregelen til
å regne ut den deriverte: $$\begin{aligned}
    \frac{d}{dx} x^2 &= (\mathbf{G}\circ \mathbf{F})'(x) = \mathbf{G}'(f(x), g(x)) \cdot \mathbf{F}'(x) \\
    &= \begin{bmatrix} g(x) & f(x) \end{bmatrix} \begin{bmatrix} f'(x) \\ g'(x)
  \end{bmatrix} = g(x) f'(x) + f(x) g'(x).    
\end{aligned}$$

## Gradienter og nivåmengder

**Definisjon 18**. En *nivåkurve* til en funksjon
$f \colon \mathbb{R}^n \to \mathbb{R}$ er en kurve
$\mathbf{r} \colon \mathbb{R}\to \mathbb{R}^n$ slik at
$f(\mathbf{r}(t)) = c$ for en konstant $c$.

**Teorem 19**. *Gitt $f \colon \mathbb{R}^n \to \mathbb{R}$ og en
nivåkurve $\mathbf{r} \colon \mathbb{R}\to \mathbb{R}^n$ til $f$ og et
tall $t$ slik at de deriverte $\mathbf{r}'(t)$ og $f'(\mathbf{r}(t))$
begge er definert. Da er
$$\nabla f(\mathbf{r}(t)) \cdot \mathbf{r}'(t) = 0.$$*

Dette er en direkte konsekvens av Oppgave
<a href="#retningsderivertderivert" data-reference-type="ref"
data-reference="retningsderivertderivert">1</a> og kjerneregelen:
$$0 = (f \circ \mathbf{r})'(t) =
  f'(\mathbf{r}(t)) \cdot \mathbf{r}'(t) =
  \nabla f(\mathbf{r}(t)) \cdot \mathbf{r}'(t).$$

Hvis $\mathbf{r} \colon \mathbb{R}\to \mathbb{R}^n$ er en hvilken som
helst deriverbar kurve, så er
$\nabla f(\mathbf{r}(t)) \cdot \mathbf{r}'(t)$ den deriverte til
$f \circ \mathbf{r}$. Det vil si at $\nabla f(\mathbf{r}(t))$ er
endringsraten til funksjonen $f$ langs kurven $\mathbf{r}$.

```python
# @title Plotter
# import matplotlib.pyplot as plt

# from matplotlib import cm
# from mpl_toolkits.mplot3d import axes3d

# Define the function
def f(x):
    x1, x2 = x
    return 2*np.exp(-((x1 + 1)**2 + x2**2)) + np.exp(-((x1 - 2)**2 + x2**2))
# Create a grid of x and y values
x1 = np.linspace(-2.5, 3.5, 300)
x2 = np.linspace(-1.5, 1.5, 300)
x = np.meshgrid(x1, x2)

# Compute the function values
x3 = f(x)

# Create a 3D plot

fig = plt.figure(figsize=plt.figaspect(1.))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.plot_surface(x[0], x[1], x3, cmap='viridis', edgecolor='none', alpha=1)
ax.set_box_aspect([6, 3, 2])
# ax = plt.figure().add_subplot(projection='3d')
ax2 = kopier_akse(fig, ax)
contour = ax2.contour3D(x1, x2, x3, alpha=1, cmap='coolwarm')  # Plot contour curves
```
